```python
import pandas as pd
```

```python
#Exportamos los archivos
pedidos=pd.read_csv('pedidos.csv',encoding='ISO-8859-1')
reseñas=pd.read_csv('reseñas.csv',encoding='ISO-8859-1')
print(pedidos)
print(reseñas)
```

```python
#Hacemos los merges necesarios para responder las preguntas
pedidos_id_in=pd.merge(pedidos, reseñas, on="pedido_id",how="inner")
pedidos_id_in
```

```python
pedidos_id_out=pd.merge(pedidos, reseñas, on="pedido_id",how="outer")
pedidos_id_out
```

```python
pedidos_id_left=pd.merge(pedidos, reseñas, on="pedido_id",how="left")
pedidos_id_left
```

```python
pedidos_id_right=pd.merge(pedidos, reseñas, on="pedido_id",how="right")
pedidos_id_right
```

```python
promedio = reseñas.groupby("producto", as_index=False).agg({
    "calificación": "mean"
})
promedio
```
