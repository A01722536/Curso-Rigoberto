```python
import pandas as pd
```

```python
trimestre_1= pd.DataFrame({"ID": [1,2,3,4],"Producto":["Bananas","Peras","Naranjas","Uvas"],"Cantidades vendidas":[500,238,345,477],"Precio por unidad":[19,22,34,40],"Ganancia":[10000,5000,9000,13000]})
trimestre_2= pd.DataFrame({"ID": [1,2,3,4],"Producto":["Bananas","Peras","Naranjas","Uvas"],"Cantidades vendidas":[360,288,306,428],"Precio por unidad":[19,22,34,40],"Ganancia":[7000,6500,9000,12000]})
trimestre_3= pd.DataFrame({"ID": [1,2,3,4],"Producto":["Bananas","Peras","Naranjas","Uvas"],"Cantidades vendidas":[430,250,322,481],"Precio por unidad":[19,22,34,40],"Ganancia":[8000,5000,9500,13500]})
```

```python
trimes=pd.concat([trimestre_1,trimestre_2,trimestre_3])
trimes
```

```python
sumado = trimes.groupby("Producto", as_index=False).agg({
    "Cantidades vendidas": "sum",
    "Ganancia": "sum"
})
sumado
```
