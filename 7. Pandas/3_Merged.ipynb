```python
import pandas as pd
```

```python
Jugadores=pd.read_csv("world_cup_players.csv")
Goleadores=pd.read_csv("world_top_scorers.csv")
print(Jugadores)
print(Goleadores)
```

```python
merged=pd.merge(Jugadores,Goleadores,on="name")
merged
```

```python
Goleadores2=pd.read_csv("world_top_scorers_2.csv")
Goleadores2
```

```python
merged2=pd.merge(Jugadores,Goleadores2,left_on="name",right_on="player_name",how="outer")
merged2
```

```python
Stats=pd.read_csv("players_stats.csv")
Stats
```

```python
Jugadores=pd.read_csv("world_cup_players.csv")
Jugadores.set_index("name",inplace=True)
Stats.set_index(Jugadores.index,inplace=True)
```

```python
merged3=pd.merge(Jugadores,Stats,left_index=True,right_index=True)
merged3
```

***Combinar DF con nombres de columnas***

```python
Jugadores=pd.read_csv("world_cup_players.csv")
Goleadores3=pd.read_csv("world_top_scorers_3.csv")
```

```python
merged4=pd.merge(Jugadores,Goleadores3,on="name",how="left",suffixes=("left","right"))
merged4
```
