```python
import pandas as pd
```

```python
clientes=pd.read_csv("clientes.csv")
ventas=pd.read_csv("ventas.csv")
print(clientes)
print(ventas)
```

```python
innered=pd.merge(clientes, ventas,on="cliente_id",how="inner")
print("INNER JOIN:")
innered
```

```python
lefted=pd.merge(clientes, ventas,on="cliente_id",how="left")
print("LEFT JOIN:")
lefted
```

```python
righted=pd.merge(clientes, ventas,on="cliente_id",how="right")
print("RIGHT JOIN:")
righted
```

```python
outered=pd.merge(clientes, ventas,on="cliente_id",how="outer")
print("OUTER JOIN:")
outered
```

**¿Qué tipo de merge mantuvo solo los clientes con compras?**
El merge con el RIGH join.

**¿En qué merge aparecen ventas sin clientes registrados?**
En los merges con el RIGH y el OUTER join.

**¿Cuál merge permite ver clientes que no han comprado nada?**
El merge con el LEFT join y el OUTER join.

**¿Cómo podrías usar merge para detectar clientes sin compras?**
Usando merges con el LEFT o el OUTER join.

**¿Cómo podrías identificar ventas sin un cliente registrado?**
Usando merges con el RIGH o el OUTER join.
