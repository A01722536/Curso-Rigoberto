```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mallorca=pd.read_excel("Mallorca2.xlsx")
mallorca
```

```python
host= mallorca[["host_id","host_name","host_since","host_is_superhost","host_listings_count"]]
listing= mallorca[["host_id","id","property_type","room_type","price"]]
print(host)
print(listing)
```

```python
df_hosts=pd.merge(host,listing, on = "host_id", how="left")
df_hosts
```

```python
df_host_count=df_hosts.groupby(["host_id","host_name","host_is_superhost"])["id"].count().reset_index()
df_host_count.rename(columns={"id":"num_listings"}, inplace=True)
df_host_count
```

```python
superhosts=df_host_count[df_host_count["host_is_superhost"]=="t"]
superhosts_sorted=superhosts.sort_values(by="num_listings",ascending=False)
print(superhosts_sorted.head(10))
```

```python
mallorca=mallorca
df_price_type=mallorca.groupby("property_type")["price"].mean().reset_index()
df_price_type.sort_values(by="price",ascending=False,inplace=True)
print(df_price_type)
```

Cuál es el barrio más caro?

```python
mallorca=mallorca.dropna(subset=["price"])
mallorca["price"]=mallorca["price"].str.replace(r"[$,]","",regex=True).astype(float)
```

```python
df_price_neigh= mallorca.groupby("neighbourhood_cleansed")["price"].mean().reset_index()
df_price_neigh.sort_values(by="price",ascending=False,inplace=True)
df_price_neigh
```

```python
price_per_room=mallorca.groupby("bedrooms")["price"].mean().reset_index()
print("Precio promedio\n",price_per_room)
```

```python
plt.figure(figsize=(10,5))#para decirle que haga la figura y el tamaño de nuestra grafica
sns.scatterplot(data= mallorca, x= 'bedrooms', y='price', alpha=0.5)
plt.title("Relacion entre numero de habitaciones y precios")
plt.xlabel("Numero de habitaciones")
plt.ylabel("Precio ($)")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,5))#para decirle que haga la figura y el tamaño de nuestra grafica
sns.histplot(data= mallorca['bedrooms'], bins=20, kde=True)
plt.title("Distribución de número de habitaciones")
plt.xlabel("Numero de habitaciones")
plt.ylabel("Frecuencia")
plt.grid()
plt.show()
```
